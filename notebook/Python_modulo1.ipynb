# ðŸ’¬ if...else

Nel codice la parola `if` viene utilizzata quando vogliamo eseguire delle istruzioni solo se una determinata condizione Ã¨ soddisfatta

Se vogliamo eseguire altri comandi quando la condizione non risulta soddisfatta, aggiungiamo la parola `else`

Nel caso in cui ci siano piÃ¹ condizioni da valutare si puÃ² usare `elif` (contrazione di â€œelse ifâ€)

Le condizioni possono essere:
* a > b
* a < b
* a == b (per l'uguaglianza devi usare il doppio uguale)
* a != b (diverso)
* a >= b
* a <= b

### ðŸ‘‰ Ora prova tu
Gioca con il valore delle variabili qui sotto e gli operatori logici e vedi come cambia il risultato

```python
numero = 4
if numero > 10:
    print("il numero Ã¨ maggiore di 10")
elif numero < 10:
    print("il numero Ã¨ minore di 10")
else:
    print("il numero Ã¨ uguale a 10")
```

```python
stringa1 = "Mi chiedi come va, come va  come va sai giÃ  come va, come va, come va "
stringa2 = "Mi chiedi come va, come va, come va, Sai giÃ  come va, come va  come va"

if stringa1 != stringa2:
    print("Le 2 stringhe sono diverse")
else:
    print("Le 2 stringhe sono uguali")
```

Puoi anche verificare se un carattere o una parola Ã¨ presente in una stringa utilizzando il costrutto `if...in`, per esempio:

```python
stringa = "CiÃ² che devi dire non l'hai detto"

if "dire" in stringa:
  print(stringa + " Tradire Ã¨ una pallottola nel petto")
```

PiÃ¹ condizioni possono essere unite con `and` e `or`, oppure negate con un not:
* condizione1 `and` condizione2 indica che entrambe le condizioni devono essere soddisfatte
* condizione1 `or` condizione2 indica che una sola delle condizioni deve essere soddisfatta
* `not` condizione indica che la condizione non deve essere soddisfatta

```python
n = 10
if n == 0:
    print('zero')
elif n == 1 or n == 2:
    print('uno o due')
elif n == 3:
    print('tre')
else:
    print('numero diverso da 0, 1, 2, 3')
```

# âš™ï¸ Funzioni

Le funzioni (o metodi) ci permettono di raggruppare un insieme di istruzioni che eseguono un compito specifico.

In Python le funzioni sono introdotte dalla parola chiave `def` seguite da un nome parlante e da una coppia di parentesi

```python
def stampa_canzone():
    print("Ãˆ difficile stare al mondo")
    print("Quando perdi l'orgoglio")
    print("Lasci casa in un giorno")

stampa_canzone()
```

Le funzioni possono accettare in input uno o piÃ¹ argomenti (o parametri), li elaborano e restituiscono in output un risultato.

Questa funzione per esempio accetta 1 valore e lo restituisce moltiplicato per 2:

```python
def raddoppia(a):
    moltiplica = a * 2
    return moltiplica

raddoppia(10)
```
